```python
# Find the mode of the "Middle Name" column
mode_value = df1['MiddleName'].mode()[0]
print(f"The mode of the 'Middle Name' column is: {mode_value}")
```

```python
# Replace null values in the "Middle Name" column with the mode
df1['MiddleName'].fillna(mode_value, inplace=True)
```

```python
null_values = df1.isnull().sum()
print(null_values)
```
